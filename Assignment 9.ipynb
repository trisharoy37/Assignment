### Explain why  we have to use the Exception class while creating a Custom Exception
we use exception class so that we can inherit the properties of the exception class, it makes it evident that your custom exception is intended to be part of the exception framework of Python. Inheriting from Exception provides semantic meaning to your custom exception. When other developers see your exception, they immediately recognize it as an error that follows the same conventions as the rest of the Python exceptions.

### Why LookupError class is used? Explain with an example KeyError and IndexError.
The LookupError class in Python is the base class for exceptions that occur when a key or index is not found during a lookup operation.

```python
#KeyError
a={"a":1,"b":2}
print(a["c"])
```

```python
#IndexError
a=[1,2,3,4,4]
print(a[10])
```
