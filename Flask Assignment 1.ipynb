### What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
In Flask, the url_for() function is used for URL building. It generates URLs for specific routes or functions defined in your Flask application.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def Jumanji():
    return 'Welcome to Jumanji!'

@app.route('/user/<username>')
def Player_Profile(username):
    return f'Player Profile page for {username}'

@app.route('/area/<int:Area_id>')
def Area_Map(Area_id):
    return f'Area Map for ID #{Area_id}'

with app.test_request_context():
    Main_url = url_for('Jumanji')
    Player_url = url_for('Player_Profile', username="Trisha_Roy")
    Area_url = url_for('Area_Map', Area_id=101)

print("Main URL:", Main_url)
print("Player Profile URL:", Player_url)
print("Area URL:", Area_url)

if __name__=="__main__":
    app.run()
```
