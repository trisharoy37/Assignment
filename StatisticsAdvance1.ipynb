Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

```python
import math
val=3
std=15
mean=100
def pdf(x):
    a=1/(std*(math.sqrt(2*3.14)))
    b=math.exp(-((1/2)*(math.pow(((x-mean)/std),2))))
    c=a*b
    return c
print(pdf(val))
```

```python
from scipy.stats import norm

def normal_pdf(x, mu, sigma):
  
  return norm.pdf(x, loc=mu, scale=sigma)

# Example usage
mean = 100
std_dev = 15
x = 3
pdf_value = normal_pdf(x, mean, std_dev)
print(f"The PDF of the normal distribution with mean {mean} and standard deviation {std_dev} at point {x} is: {pdf_value}")
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.

Ans- Properties of Binomial Distributions
1. Every experiment outcome is binary.
2. These experiment is performed for n trials and number of trials is fixed.
3. The probability of success is constant for each trial.
4. The trials are independent.
5. Applicable on discrete random variables only.

Example:
1. Counting the number of individuals who own more than one car.
2. Counting the number of correct answers in a multi-choice exam.




Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
and plot a histogram of the results using matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define parameters
n = 1000  # Number of trials
p = 0.4  # Probability of success

# Generate random sample
samples = np.random.binomial(n, p, size=1000)

# Create histogram with informative labels and title
plt.hist(samples, bins=20, edgecolor='black', label=f"Binomial Distribution (n={n}, p={p})")
plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.title('Histogram of Binomial Distribution')
plt.legend()
plt.grid(True)

# Improve code clarity and formatting
plt.tight_layout()  # Adjust spacing to prevent overlapping elements
plt.show()
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

```python
import scipy.special as sp
def poisson_cdf(k,l):
    if k<0:
        return 0
    else:
        return sp.gammainc(k+1,l)
mean=6
x=9
cdf_value=poisson_cdf(x, mean)
print(cdf_value)
```

Q7. How Binomial distribution different from Poisson distribution?

1. Binomial:
    * Fixed number of trials. Each trial has only two outcome: success or failures.
    * Constant probability of success(p) for each trial. This means the chance of heads in a coin toss stays the same throughout all flips. It calculates the probability of getting a specific number of successes (k) within those fixed trials.
2. Poisson:
    * Events happening within a fixed interval(like customer arrivals in an hour). The number of events can be zero, one, or any positive integer (unlike the fixed two in binomial).
    * Average rate (lambda) at which events occur within the interval. This represents the expected number of events, but the actual number can vary. It calculates the probability of observing a certain number of events (k) within the fixed interval


Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
sample mean and variance.

```python
import numpy as np

sample_size=1000
mean = 5

samples=np.random.poisson(mean, size=sample_size)

sample_mean=np.mean(samples)
sample_variance=np.var(samples)

print(f"Sample mean: {sample_mean:.2f}")
print(f"Sample variance: {sample_variance:.2f}")
```

Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

In Binomial distribution mean = n x p, whereas variance= n x p x (1-p).
Here,
If mean increases, the variance also increases proportionally.
However, the variance is maximized when the probability of success(p) is around 0.5. As p approaches 0 or 1, the variance decreases.

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

In a normal distribution, the least frequent data appears in the tails of the distribution, further away from the mean.  The normal distribution is symmetrical around the mean, so the data is less frequent the farther you move in either direction (left or right) from the center.
