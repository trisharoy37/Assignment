Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

```python
import math
val=eval(input("Enter the value of point"))
std=15
mean=100
def pdf(x):
    a=1/(std*(math.sqrt(2*3.14)))
    b=math.exp(-((1/2)*(math.pow(((x-mean)/std),2))))
    c=a*b
    return c
print(pdf(val))
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.

Ans- Properties of Binomial Distributions
1. Every experiment outcome is binary.
2. These experiment is performed for n trials and number of trials is fixed.
3. The probability of success is constant for each trial.
4. The trials are independent.
5. Applicable on discrete random variables only.

Example:
1. Counting the number of individuals who own more than one car.
2. Counting the number of correct answers in a multi-choice exam.




```python
import matplotlib.pyplot as plt
import numpy as np

# Define parameters
n = 1000  # Number of trials
p = 0.4  # Probability of success

# Generate random sample
samples = np.random.binomial(n, p, size=1000)

# Create histogram with informative labels and title
plt.hist(samples, bins=20, edgecolor='black', label=f"Binomial Distribution (n={n}, p={p})")
plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.title('Histogram of Binomial Distribution')
plt.legend()
plt.grid(True)

# Improve code clarity and formatting
plt.tight_layout()  # Adjust spacing to prevent overlapping elements
plt.show()
```
